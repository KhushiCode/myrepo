## The effectiveness parameter of the tweets have been calculated as follows:
1. The time_from_death has been given a weightage of 20% since the tweets didn't have much time difference at the time of tweeting.
2. The presence of video and retweet counts have been given a weightage of 30% because the tweet is really highly effective if it has been retweeted high number of times.
3. The feature likes_count have has been given a weightage of 20% because the tweet with high number of likes is obviously effective.

The above understandings have been taken from the official declaration of the twitter.


Then, the above values have been scaled to 0 to 1 using sigmoid function. 

image.png

Then scaled values have all been added up to obtain the effectiveness score of the tweet.

```python
df_effective=df[["username", "retweets_count", "likes_count", "video", "time_from_death_minutes"]]
df_effective.head()
```

```python
df_effective.shape
```

```python
df_effective.describe()
```

```python
df_effective["time_from_death_minutes"]=df_effective["time_from_death_minutes"]*0.2
df_effective["video"]=df_effective["video"]*0.3
df_effective["likes_count"]=df_effective["likes_count"]*0.2
df_effective["retweets_count"]=df_effective["retweets_count"]*0.3
```

```python
import numpy as np



for i in tqdm(range(len(df_effective))):
  df_effective["time_from_death_minutes"][i]=1/(1 + np.exp(-df_effective["time_from_death_minutes"][i]))
  df_effective["video"][i]=1/(1 + np.exp(-df_effective["video"][i]))
  df_effective["retweets_count"][i]=1/(1 + np.exp(-df_effective["retweets_count"][i]))
  df_effective["likes_count"][i]=1/(1 + np.exp(-df_effective["likes_count"][i]))

df_effective["effective_score"]=df_effective["time_from_death_minutes"]+df_effective["video"]+df_effective["retweets_count"]+df_effective["likes_count"]
```
